## Plottling LRs

```python
colors_f1_score = {'f1score_train_LR1': 'tab:grey', 'f1score_val_LR1': 'tab:grey', 
                   'f1score_train_LR2': 'tab:green', 'f1score_val_LR2': 'tab:green',
                   'f1score_train_LR3': 'tab:orange', 'f1score_val_LR3': 'tab:orange',
                   'f1score_train_LR4': 'tab:blue', 'f1score_val_LR4': 'tab:blue',
                   'f1score_train_LR5': 'tab:red', 'f1score_val_LR5': 'tab:red'
                  }

colors_loss = {'loss_train_LR1': 'tab:purple', 'loss_val_LR1': 'tab:purple',  
               'loss_train_LR2': 'peru', 'loss_val_LR2': 'peru',
               'loss_train_LR3': 'steelblue', 'loss_val_LR3': 'steelblue',
               'loss_train_LR4': 'yellowgreen', 'loss_val_LR4': 'yellowgreen',
               'loss_train_LR5': 'firebrick', 'loss_val_LR5': 'firebrick'
              }
```

```python
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, hspace=0, wspace=0)
ax1 = gs.subplots(sharex='col', sharey='row')

fig.suptitle(f'Train and Validation - F1 Score and Loss Plots', fontsize=16)

epochs = [i for i in range(int(num_epochs))]

#plot = plot_names[i]
#ax1[i].set_title(f'{plot}')
for j in range(5):

    ax1[0][0].plot(epochs, metric_params['f1_score'][f'f1score_train_LR{j+1}'],
                    label = f'LR = {l_rates[j]}-F1 Score (Train)', 
                    color = colors_f1_score[f'f1score_train_LR{j+1}']
                    )
    ax1[0][0].set_xlabel('Loss Plot')
    ax1[0][0].set_ylabel('F1 Score')
    ax1[0][0].set_title("Training")
    #ax1[0][0].legend(bbox_to_anchor=(0.5, -1.05, 0.40, 0.94))

    ax1[0][1].plot(epochs, metric_params['f1_score'][f'f1score_val_LR{j+1}'],
                    label = f'LR = {l_rates[j]}-F1 Score (Val)', 
                    color = colors_f1_score[f'f1score_val_LR{j+1}']
                    )
    ax1[0][1].set_title("Validation")
    #ax1[0][1].legend(bbox_to_anchor=(0.5, -1.05, 0.40, 0.94))


for j in range(5):
    ax1[1][0].plot(epochs, np.log(metric_params['loss'][f'loss_train_LR{j+1}']),
                    label = f'LR = {l_rates[j]}-Loss (Train)', 
                    color = colors_loss[f'loss_train_LR{j+1}']
                    )
    ax1[1][0].set_ylabel('Loss (Log Scale)')
    ax1[1][0].set_xlabel('Epochs')
    #ax1[1][0].legend(bbox_to_anchor=(0.5, -1.05, 0.40, 0.94))

    ax1[1][1].plot(epochs, np.log(metric_params['loss'][f'loss_val_LR{j+1}']),
                    label = f'LR = {l_rates[j]}-Loss (Val)', 
                    color = colors_loss[f'loss_val_LR{j+1}']
                    )
    ax1[1][1].set_xlabel('Epochs')
    #ax1[1][1].legend(bbox_to_anchor=(0.5, -1.05, 0.40, 0.94))
    #ax1[1][1].legend(bbox_to_anchor=(0, -1.05, 0.40, 0.94))
        
    # for k in range(2):
    #     handles, labels = ax1[i][k].get_legend_handles_labels()
    #     hand.append(handles)
    #     lab.append(labels)
        

# handles_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
# handles, labels = [sum(handles_and_labels, []) for handles_and_labels in zip(*lines_labels)]
# plt.legend(handles, labels, bbox_to_anchor=(0.5, -1.05, 0.40, 0.94))

ax1[0][0].legend(bbox_to_anchor=(0.4, -1.15, 0, 0))
ax1[1][0].legend(bbox_to_anchor=(0.9, -0.15, 0, 0))
ax1[0][1].legend(bbox_to_anchor=(0.4, -1.15, 0, 0))
ax1[1][1].legend(bbox_to_anchor=(0.9, -0.15, 0, 0))

ax1[1][0].set_ylim(-10, 20)
ax1[1][1].set_ylim(-10, 20)
fig.tight_layout()
fig.savefig("ablation.eps", format="eps", dpi=1200, bbox_inches="tight", transparent=True)
```
