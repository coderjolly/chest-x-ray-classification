## Covid 19, Viral and Normal Images

```python
os.getcwd()
```

```python
#Viral Pneumonia
im1 = read_image("Viral Pneumonia-635.png", mode=torchvision.io.ImageReadMode.GRAY)
```

```python
#Covid 19
im2 = read_image("COVID-554.png", mode=torchvision.io.ImageReadMode.GRAY)
```

```python
#Normal
im3 = read_image("Normal-762.png", mode=torchvision.io.ImageReadMode.GRAY)
```

```python
fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(1, 3, hspace=0, wspace=0)
ax1, ax2, ax3 = gs.subplots(sharex='col', sharey='row')

ax1.imshow(np.transpose(im2.numpy(), (1, 2, 0)), cmap = 'gray')
ax2.imshow(np.transpose(im1.numpy(), (1, 2, 0)), cmap = 'gray')
ax3.imshow(np.transpose(im3.numpy(), (1, 2, 0)), cmap = 'gray')

ax1.title.set_text('Covid-19')
ax2.title.set_text('Viral Pneumonia')
ax3.title.set_text('Normal')

ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_yticklabels([])
ax1.set_xticklabels([])

ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_yticklabels([])
ax2.set_xticklabels([])

ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_yticklabels([])
ax3.set_xticklabels([])

ax1.patch.set_edgecolor('black')  
ax1.patch.set_linewidth('1')  
fig.savefig("comparison_of_covid_pneumonia_normal.eps", format="eps", dpi=1200, bbox_inches="tight", transparent=True)
```
