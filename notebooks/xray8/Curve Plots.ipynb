```python
fig, ax1 = plt.subplots()
plt.title('Validation Set Accuracy & Loss Plot')
epochs = [i for i in range(int(num_epochs))]

# plt.figure()
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
color = 'tab:green'
for model, vals in val_acc.items():
    plt.plot(epochs, 
             vals,
             label=f'Model: {model}')
ax1.tick_params(axis='y', labelcolor=color)   

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Loss', color=color)


for model, vals in val_loss.items():
    plt.plot(epochs, 
             np.log(vals),
             label=f'Model: {model}')
ax2.tick_params(axis='y', labelcolor=color)    

fig.tight_layout()
fig.legend()
fig.show()
```

```python
colors_acc = {'chest_xray_8': 'yellowgreen', 'covid_pneumonia': 'lightcoral', 'pneumonia': 'cornflowerblue'}
colors_loss = {'chest_xray_8': 'green', 'covid_pneumonia': 'firebrick', 'pneumonia': 'blue'}
middle_flag = {'resnet34': False, 'efficient_net_b1': True, 'mobile_net_v3_large': False}
start_flag = {'resnet34': True, 'efficient_net_b1': False, 'mobile_net_v3_large': False}
end_flag = {'resnet34': False, 'efficient_net_b1': False, 'mobile_net_v3_large': True}
```

```python
model_names = {'resnet34': "ResNet 34", 
               'mobile_net_v3_large': "Mobile Net V3 Large", 
               'efficient_net_b1': "Efficient Net B1"}
dataset_names = {'chest_xray_8': "Chest X-Ray 8", 
               'covid_pneumonia': "COVID + Pneumonia", 
               'pneumonia': "Pneumonia"}
```

```python
#fig, ax1 = plt.subplots(1,3, figsize=(12, 5))
fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(1, 3, hspace=0, wspace=0)
ax1 = gs.subplots(sharex='col', sharey='row')


ax2 = []
fig.suptitle(f'Train & Validation Dataset - Accuracy & Loss Plots', fontsize=16)
epochs = [i for i in range(int(num_epochs))]
for i in range(3):
    model = model_name[i]
    ax1[i].set_title(f'{model_names[model]}')
    # plt.figure()
    if start_flag[model]:
        ax1[i].set_ylabel('Accuracy')
    if middle_flag[model]:
        ax1[i].set_xlabel('Epoch')
    
    color = 'black'

    for data in model_params.keys():
        ax1[i].plot(epochs, 
                 model_params[data]['val_acc'][model],
                 label=f'{dataset_names[data]} - Validation Accuracy', 
                 color=colors_acc[data])
        ax1[i].plot(epochs, 
                 model_params[data]['train_acc'][model],
                 label=f'{dataset_names[data]} - Training Accuracy', 
                 color=colors_acc[data], linestyle='--')

    ax1[i].tick_params(axis='y', labelcolor=color)   
    ax1[i].set_ylim([0, 1.05])

    ax2.append(ax1[i].twinx())
    color = 'black'
    
    if end_flag[model]:
        ax2[i].set_ylabel('Loss (Log Scale)', color=color)

    for data in model_params.keys():
        ax2[i].plot(epochs, 
                 np.log(model_params[data]['val_loss'][model]),
                 label=f'{dataset_names[data]} - Validation Loss', 
                 color=colors_loss[data])
        ax2[i].plot(epochs, 
                 np.log(model_params[data]['train_loss'][model]),
                 label=f'{dataset_names[data]} - Training Loss', 
                 color=colors_loss[data], linestyle='--')

    ax2[i].set_ylim([-8, 20])
    ax2[i].tick_params(axis='y', labelcolor=color)    
    
    if start_flag[model]:
        ax2[i].set_yticklabels([])
        ax1[i].legend(bbox_to_anchor=(2., -1.05, 0.41, 0.94))
        ax2[i].legend(bbox_to_anchor=(1., -1.05, 0.41, 0.94))
        #ax1[i].get_legend().remove()
        #ax2[i].get_legend().remove()
    if middle_flag[model]:
        #ax1[i].set_yticklabels([])
        ax2[i].set_yticklabels([])
        #ax1[i].get_legend().remove()
        #ax2[i].get_legend().remove()

        

fig.tight_layout()
fig.savefig("acc_loss.png", format="png", dpi=600, bbox_inches="tight")
fig.savefig("acc_loss.eps", format="eps", dpi=1200, bbox_inches="tight", transparent=True)
fig.savefig("acc_loss.pdf", format="pdf", dpi=1200, bbox_inches="tight", transparent=True)

#fig.legend(bbox_to_anchor=(0., -1., 0.41, 0.94))
```
